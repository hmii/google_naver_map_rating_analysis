```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
from IPython.display import display
import time
```

# 애월읍 음식점 스크래핑

```python
driver = webdriver.Chrome('./chromedriver')
driver.implicitly_wait(10)


# 애월읍 음식점 
aewol = ['고내리', '고성리', '곽지리', '광령리', '구엄리', '금성리', '납읍리', '봉성리', '상가리', '상귀리', '소길리', '수산리', '신엄리', '애월리', '어음리', '유수암리', '장전리', '하가리', '하귀1리', '하귀2리']

# 성산읍 음식점 
# sungsan = ['고성리', '난산리', '삼달리', '성산리', '수산리', '시흥리', '신산리', '신천리', '신풍리', '오조리', '온평리']

eup = '애월읍'

for ri in aewol : 
# for ri in sungsan :

    url = 'https://www.google.com/maps/search/애월읍+{}+음식점/@33.4767855,126.3215268,14z/data=!3m1!4b1'.format(ri)
#     url = 'https://www.google.com/maps/search/성산읍+{}+음식점/@33.4767855,126.3215268,14z/data=!3m1!4b1'.format(ri)
    driver.get(url)
    driver.implicitly_wait(10)

    data = []

    for _ in range(100) :

        time.sleep(2)

        try : 
            for i in range(20):
                scroll = WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.XPATH, '//*[@id="pane"]/div/div[1]/div/div/div[4]/div[1]')))
                driver.execute_script("arguments[0].scrollBy(0,arguments[0].scrollHeight)", scroll)  
        except :
            break

        time.sleep(3)


        # 소스를 읽어 숩에서 파싱
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        olle = soup.select('#pane > div > div.widget-pane-content.cYB2Ge-oHo7ed > div > div > div.section-layout.section-scrollbox.cYB2Ge-oHo7ed.cYB2Ge-ti6hGc.siAUzd-neVct-Q3DXx-BvBYQ > div.section-layout.section-scrollbox.cYB2Ge-oHo7ed.cYB2Ge-ti6hGc.siAUzd-neVct-Q3DXx-BvBYQ > div')


        for store in olle[::2]:
            try:
                name = store.select("span")[0].text.strip() 
            
                try :
                    point = store.select('span.ZkP5Je')[0].text
                    score = point[:3]
                    review = point[4:-1]
                except :
                    score = ''
                    review= ''
                try :
                    what = store.select('span:nth-child(1) > jsl > span:nth-child(2)')[0].text
                except :
                    what = ''
                try:
                    address = store.select('span:nth-child(2) > jsl > span:nth-child(2)')[0].text
                    address = address.replace("KR", "")
                    address = address.replace("제주특별자치도", "")
                    address = address.replace("특별자치도", "").strip()
                except :
                    address = ''
                try:
                    price = store.select('span:nth-child(3) > jsl ')[0].text
                    price = price.replace("·", "").strip()
                except:
                    price = ''
                data.append([name, score, review, price, what, address])
                
            except:
                pass 

        # 다음 페이지 클릭 

        try:
            driver.find_element_by_id('ppdPk-Ej1Yeb-LgbsSe-tJiF1e').click()
            time.sleep(2)
        except:
            break

    
    df = pd.DataFrame(data, columns = ['name', 'score', 'review', 'price', 'what', 'address']) 
    print(f'{eup} {ri}의 데이터는 {df.name.count()}개 입니다')
    df.to_excel(f'df_{eup}_{ri}.xlsx', index=False)
#     df2 = df.drop_duplicates()
#     display(df2.count())
#     df2.to_excel(f'df_drop_{ri}.xlsx', index=False) 
driver.close()
```
