# 주소 걸러내기

```python
# aewol = ['고내리', '고성리', '곽지리', '광령리', '구엄리', '금성리', '납읍리', '봉성리', '상가리', '상귀리', '소길리', '수산리', '신엄리', '애월리', '어음리', '유수암리', '장전리', '하가리', '하귀1리', '하귀2리']
aewol = ['고성리', '난산리', '삼달리', '성산리', '수산리', '시흥리', '신산리', '신천리', '신풍리', '오조리', '온평리']

print(aewol)
```

```python
# 애월읍 리 합치고 '리' 없애기 
result = "|".join(aewol)
result = result.replace("리", "")
result
```

```python
# 정확성을 위해 다른 키워드도 추가 
addr = result+'|성산|일출|동류암'
addr
```

```python
# 해당 주소 키워드를 포함하는 주소를 가진 행만 뽑아내기 
aw2 = aw[aw['address'].str.contains(addr, regex=True, na=False)] # nan일 때 false반환 
aw2
```

```python
# 해당 주소 키워드를 가지지 않은 리스트 따로 뽑아서 변수에 저장
not_aw = aw[~aw['address'].str.contains(addr, regex=True, na=False)]
not_aw.address.unique()
```

```python
# 애월읍이 아닌 주소 뽑아낸 뒤 혹시 잘못빠진건 아닌지 확인  
not_aw.address.unique()
```

```python
# 파일로 저장 
not_aw.to_excel('애월읍_이외.xlsx', index=False)
```

```python
aw4.what.unique()   # 어떤 업종을 포함하는지 확인 
```

```python
# 정리된 파일로 저장 
aw4.to_excel('애월읍_음식점만.xlsx', index=False)   
```
